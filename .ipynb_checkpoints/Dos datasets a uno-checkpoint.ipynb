```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import dfply
import plotly
from scipy import stats
%matplotlib inline
```

```python
data1 = pd.read_csv('UFO_tidy.csv', sep = ",", engine = 'python')
data2 = pd.read_csv('nuclear_tidy.csv', sep = ",", engine = 'python', index_col = None)
```

```python
plant_in_city = []
plant_name = []
plant_latitude = []
plant_longitude = []

for row in data1['City'] :
    
    if row in data2.City.unique() :
        
        data = data2 >> dfply.filter_by(data2['City'] == row)
        
        plant = 'Yes'
        name = data['Plant']
        latitude = data['Latitude']
        longitude = data['Longitude']
        
        plant_in_city.append(plant)
        plant_name.append(name)
        plant_latitude.append(latitude)
        plant_longitude.append(longitude)
            
        
        print(row, plant, name)
    
    else :
        
        plant = 'No'
        name = 'None'
        latitude = 'None'
        longitude = 'None'
        
        plant_in_city.append(plant)
        plant_name.append(name)
        plant_latitude.append(latitude)
        plant_longitude.append(longitude)
        
        
        print(row, plant, name)
```

```python
data1['plant_in_city'] = plant_in_city
data1['plant_name'] = plant_name
data1['plant_latitude'] = plant_latitude
data1['plant_longitude'] = plant_longitude
```

```python
data1.plant_in_city.value_counts()
```

```python
data1.to_csv('UFO-nuclear.csv', index = False)
```

# Plantas nucleares en el Reino Unido

```python
city2 = []
state2 = []
country2 = []
address_2 = []

for row in data2['latlong'].values:
    
    resultado = geolocator.reverse(row)

    s = str(resultado)
    s = s.split(',')
    
    city = s[1]
    state = s[-3]
    country = s[-1]
    address = s
    
    city2.append(city)
    state2.append(state)
    country2.append(country)
    address_2.append(address)
    
    print(city, state, country)
```

```python
data2['city'] = city2
data2['state'] = state2
data2['Country'] = country2
data2['address'] = address_2
```

```python
data2.head()
```

```python
data2.to_csv('nuclear_geo.csv', index = False)
```

# Avistamientos en el Reino Unido

El método más efectivo es el de Nominatim, sin embargo, este contiene limitaciones sobre la cantidad de datos que se pueden accesar. Para trabajar con ello, se dividirá el dataset en subgrupos y despué se reagrupará.

```python
data1['shape'].value_counts()
```

```python
sub1_1 = data1 >> dfply.mask(data1['shape'] == 'light')
sub1_2 = data1 >> dfply.mask(data1['shape'] == 'circle')
sub1_3 = data1 >> dfply.mask(data1['shape'] == 'triangle')
```

```python
subs = [sub1_1, sub1_2, sub1_3]
sub1 = pd.concat(subs)
```

```python
sub1d = sub1.index
sub2 = data1.drop(sub1d)
```

```python
sub1.shape
```

```python
sub2.shape
```

#### Primera parte:

```python
city1 = []
state1 = []
country1 = []
address_l = []

for row in sub1['latlong'].values:
    
    resultado = geolocator.reverse(row)

    s = str(resultado)
    s = s.split(',')
    
    city = s[1]
    state = s[-3]
    country = s[-1]
    address = s
    
    city1.append(city)
    state1.append(state)
    country1.append(country)
    address_l.append(address)
    
    print(city, state, country)
```

```python
sub1['City'] = city1
sub1['State'] = state1
sub1['Country'] = country1
sub1['Address'] = address_l
```
